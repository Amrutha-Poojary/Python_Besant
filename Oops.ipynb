OOPS

Object Oriented Programming (OOP) is a programming paradigm that uses "objects" to design applications and computer programs.

Object: Real-world entity that exists in reality  
Orientation: Perception  

In Python:
1. Everything is an object  
2. Every object belongs to a particular class  
3. Every object has two parts  
4. Everything is useful  
5. Every object is in constant interaction  

Has Part                  Does Part
--------                  ---------
Properties                Actions
Variables                 Methods / Functions

Example:

Has                       Does
---                       ----
Door                      Open()
Wheel                     Rotate()
Steering                  TurnLeft()
Engine                    Start()
Brake                     Stop()

```python
class Car:

    def __init__(self, name,yom,clr,price,fuel,max_speed):
        self.name=name
        self.yom=yom
        self.clr=clr
        self.price=price
        self.fuel=fuel
        self.max_speed=max_speed

    def dispaly_details(self):
        print(f': {self.name}')
        print(f': {self.clr}')
        print(f': {self.price}')

    def is_supercar(self):
        if self.max_speed > 200:
            print("is supercar")
        else:
            print("is not a supercar")
```

```python
bmw = Car('bmw 800', 2019, 'black', 9000000,'petrol',350)
print(bmw.name)
print(bmw.price)

bmw.dispaly_details()
bmw.is_supercar()
```
