Types of arguments



    1.positional

```python
def power(a,b):
    return a**b

print(power(10,2))
print(power(2,10))
```

    2.keyword argument

```python
def power(a,b):
    return a**b

print(power(b=10,a=2))
print(power(a=2,b=10))
```

```python
def add_details(name, order, price):
    print(name, order, price)
add_details('kiran', 'ice cream',40)
```

    3.variable length argument
    

```python
def get_details(*details):
    print(details)
get_details()
get_details('kiran')
get_details('kiran','ice cream')
```

    4.variable keyword argument

```python
def get_details(**details):
    print(details)
get_details()
get_details(name='kiran')
get_details(name='kiran',order='ice cream')
```

```python
def get_cus_details(a,*args,**kwargs):   # order is positional, variable length, variable keyword 
    print(a,args,kwargs)
get_cus_details(10, 'kiran', name='kiran')
```

```python
# this is recursive function

def fact(n):
    if n==0:
        return 1
    else:
        return n*fact(n-1)
fact(5)
```

```python
def fact(n):
    if n==0:
        return 1
    else:
        res = 1
        while n>0:
            res =res*n
            n=n-1
            print(res,n)
        return res
fact(5)
```
