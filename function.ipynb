Types of arguments



    1.positional

```python
def power(a,b):
    return a**b

print(power(10,2))
print(power(2,10))
```

    2.keyword argument

```python
def power(a,b):
    return a**b

print(power(b=10,a=2))
print(power(a=2,b=10))
```
